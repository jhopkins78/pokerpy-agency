```python
import re
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from datasets import load_dataset
from tqdm.auto import tqdm
test_dataset = dataset["test"]

device = 0 if torch.cuda.is_available() else -1
text_gen = pipeline(
    "text-generation",
    model="SoAnVa/my_awesome_poker_model",
    tokenizer="SoAnVa/my_awesome_poker_model",
    return_full_text=False,
    device=device
)


predictions = []
ground_truths = []

# Loop over test examples
for example in tqdm(test_dataset, desc="Processing test examples"):
    prompt = example["question"]
    true_answer = example["answer"]
    
    # Generate answer using the pipeline (do_sample=False for deterministic output)
    result = text_gen(prompt, max_new_tokens=10, do_sample=False)
    generated_text = result[0]["generated_text"]
    
    # Extract only the action keyword from both ground truth and generated answer.
    true_action = extract_action(true_answer)
    pred_action = extract_action(generated_text)
    
    ground_truths.append(true_action)
    predictions.append(pred_action)
```

```python
# Define the list of labels (including "NONE" for no action detected)
labels = ["FOLD", "CALL", "CHECK", "BET", "ALLIN", "RAISE", "NONE"]

# Create a confusion matrix.
cm = confusion_matrix(ground_truths, predictions, labels=labels)
cm_df = pd.DataFrame(cm, index=labels, columns=labels)

plt.figure(figsize=(10, 8))
sns.heatmap(cm_df, annot=True, fmt='d', cmap="Blues")
plt.xlabel("Predicted Action")
plt.ylabel("True Action")
plt.title("Confusion Matrix of Poker Actions")
plt.show()

print("Classification Report:")
print(classification_report(ground_truths, predictions, labels=labels))
```

```python
import re
import numpy as np
import torch
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define ranks in descending order (Ace high, 2 low)
ranks = ["A", "K", "Q", "J", "T", "9", "8", "7", "6", "5", "4", "3", "2"]

# Synthetic prompt template with a placeholder for the hole cards.
# Here, P1 is acting. With BTN=P5, SB=P1, BB=P2, P1 is UTG.
prompt_template = """Question: [TABLE_CONFIGURATION]
BTN=P5
SB=P1 0.5BB
BB=P2 1BB

[STACKS]
P1: 20.8BB {hole_cards}
P2: 23.4BB 
P3: 32.7BB
P4: 83.2BB
P5: 56.1BB
POT=1.5BB

[PREFLOP]
P1:"""

# Compile the action pattern.
# Only valid actions: FOLD, CALL, CHECK, BET, ALLIN, RAISE.
action_pattern = re.compile(r'\b(FOLD|CALL|CHECK|BET|ALLIN|RAISE)\b', re.IGNORECASE)
def extract_action(text):
    """Extracts the first occurrence of a valid action in uppercase or returns 'NONE'."""
    match = action_pattern.search(text)
    return match.group(1).upper() if match else "NONE"

# Create a text-generation pipeline.
# Using return_full_text=False so that the generated text does not include the prompt.
device = 0 if torch.cuda.is_available() else -1
text_gen = pipeline(
    "text-generation",
    model="SoAnVa/my_awesome_poker_model",
    tokenizer="SoAnVa/my_awesome_poker_model",
    return_full_text=False,
    device=device
)

# We'll only count actions FOLD, RAISE, and CHECK.
action_interest = ["FOLD", "RAISE", "CHECK"]

# Initialize three 13x13 matrices (one per action) to hold percentages.
fold_matrix  = np.full((13, 13), np.nan)
raise_matrix = np.full((13, 13), np.nan)
check_matrix = np.full((13, 13), np.nan)

# Define suit groups for unsuited hands.
red_suits = ["h", "d"]   # hearts and diamonds
black_suits = ["c", "s"] # clubs and spades

# Loop over all combinations of starting hands.
# Convention: rows (i) and columns (j) correspond to the two hole card ranks.
# Upper triangle (i < j): suited hands.
# Diagonal (i == j): pairs.
# Lower triangle (i > j): offsuit hands.
for i in range(13):
    for j in range(13):
        simulation_hands = []  # list of hole_cards strings to simulate
        if i == j:
            # For pairs, choose two different suits (e.g., clubs and diamonds)
            card1 = ranks[i] + "c"
            card2 = ranks[j] + "d"
            simulation_hands.append(f"[{card1} {card2}]")
        elif i < j:
            # Suited hands: choose same suit (e.g., clubs)
            card1 = ranks[i] + "c"
            card2 = ranks[j] + "c"
            simulation_hands.append(f"[{card1} {card2}]")
        else:
            # Offsuit: simulate multiple combinations.
            unsuited_hands = []
            # Option 1: first card red, second card black.
            for s1 in red_suits:
                for s2 in black_suits:
                    unsuited_hands.append(f"[{ranks[i]}{s1} {ranks[j]}{s2}]")
            # Option 2: first card black, second card red.
            for s1 in black_suits:
                for s2 in red_suits:
                    unsuited_hands.append(f"[{ranks[i]}{s1} {ranks[j]}{s2}]")
            simulation_hands.extend(unsuited_hands)
        
        # Counters for actions of interest.
        counts = {act: 0 for act in action_interest}
        total_simulations = 0
        
        # For each simulation (each distinct hole card combination), generate a model decision.
        for hole_cards in simulation_hands:
            prompt = prompt_template.format(hole_cards=hole_cards)
            result = text_gen(prompt, max_new_tokens=10, do_sample=True, top_k=50, top_p=0.95, temperature=0.5)
            generated_text = result[0]["generated_text"]
            action = extract_action(generated_text)
            # Only count if the extracted action is in our interest list.
            if action in action_interest:
                counts[action] += 1
            total_simulations += 1
        
        # Compute percentages if any simulation was run.
        if total_simulations > 0:
            fold_pct  = (counts["FOLD"]  / total_simulations) * 100
            raise_pct = (counts["RAISE"] / total_simulations) * 100
            check_pct = (counts["CHECK"] / total_simulations) * 100
        else:
            fold_pct = raise_pct = check_pct = np.nan
        
        fold_matrix[i, j]  = fold_pct
        raise_matrix[i, j] = raise_pct
        check_matrix[i, j] = check_pct

# Function to plot a heatmap.
rank_labels = ranks  # use the ranks as labels for both axes
def plot_heatmap(matrix, title):
    df = pd.DataFrame(matrix, index=rank_labels, columns=rank_labels)
    plt.figure(figsize=(10, 8))
    sns.heatmap(df, annot=True, fmt=".1f", cmap="viridis", cbar_kws={'label': '%'})
    plt.xlabel("Second Card Rank")
    plt.ylabel("First Card Rank")
    plt.title(title)
    plt.show()

plot_heatmap(fold_matrix, "FOLD Percentage")
plot_heatmap(raise_matrix, "RAISE Percentage")
plot_heatmap(check_matrix, "CHECK Percentage")
```

```python
import re
import numpy as np
import torch
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
from datasets import load_dataset
from tqdm.auto import tqdm
import matplotlib.pyplot as plt

# Adjusted extraction function to capture action and its bet amount (in BB)
def extract_action_amount(text):
    """
    Extract the first occurrence of one of the valid actions (FOLD, CALL, CHECK, BET, ALLIN, RAISE)
    and capture the numeric bet amount (if present) from the text.
    Returns a tuple: (action, amount). If no amount is found, amount is 0.0.
    """
    pattern = re.compile(r'\b(FOLD|CALL|CHECK|BET|ALLIN|RAISE)\b\s*(\d+(?:\.\d+)?)?\s*BB?', re.IGNORECASE)
    match = pattern.search(text)
    if match:
        action = match.group(1).upper()
        amount_str = match.group(2)
        amount = float(amount_str) if amount_str is not None else 0.0
        return action, amount
    else:
        return "NONE", 0.0

# Load your test dataset (here, we select the first 200 examples)
dataset = load_dataset("SoAnVa/private_processed_demo")
test_dataset = dataset["test"]

# Create a text-generation pipeline with do_sample=False for deterministic output.
device = 0 if torch.cuda.is_available() else -1
text_gen = pipeline(
    "text-generation",
    model="SoAnVa/my_awesome_poker_model",
    tokenizer="SoAnVa/my_awesome_poker_model",
    return_full_text=False,
    device=device
)

# We are interested in actions that come with a bet amount.
actions_with_amount = ["CALL", "BET", "ALLIN", "RAISE"]

# Create a dictionary to store pairs of true and predicted amounts for each action.
pairs = {act: {"true": [], "pred": []} for act in actions_with_amount}

# Loop over test examples.
for example in tqdm(test_dataset, desc="Processing test examples"):
    prompt = example["question"]
    true_text = example["answer"]
    
    # Generate response using the pipeline.
    result = text_gen(prompt, max_new_tokens=10, do_sample=False)
    generated_text = result[0]["generated_text"]
    
    # Extract (action, amount) from both ground truth and generated answer.
    true_action, true_amount = extract_action_amount(true_text)
    pred_action, pred_amount = extract_action_amount(generated_text)
    
    # For actions of interest, if the predicted action matches the ground truth, record the pair.
    if true_action in actions_with_amount and pred_action == true_action:
        pairs[true_action]["true"].append(true_amount)
        pairs[true_action]["pred"].append(pred_amount)

# Now, for each action, plot a scatter plot comparing the true and predicted bet amounts.
n_actions = len(actions_with_amount)
fig, axes = plt.subplots(n_actions, 1, figsize=(8, n_actions * 4), sharex=True)

for idx, act in enumerate(actions_with_amount):
    ax = axes[idx]
    true_vals = pairs[act]["true"]
    pred_vals = pairs[act]["pred"]
    
    if true_vals and pred_vals:
        ax.scatter(true_vals, pred_vals, alpha=0.7)
        # Compute min and max for plotting the y=x line
        combined = true_vals + pred_vals
        xmin, xmax = min(combined), max(combined)
        ax.plot([xmin, xmax], [xmin, xmax], 'r--', label='Perfect Match (y=x)')
        ax.set_title(f"{act}: True vs. Predicted Amounts")
        ax.set_ylabel("Predicted Amount (BB)")
        ax.legend()
    else:
        ax.text(0.5, 0.5, f"No data for {act}", horizontalalignment='center', verticalalignment='center')
        ax.set_title(f"{act}: True vs. Predicted Amounts")
        ax.set_ylabel("Predicted Amount (BB)")
    
ax.set_xlabel("True Amount (BB)")
plt.tight_layout()
plt.show()
```
